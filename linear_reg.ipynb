```python
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
# all class converted into numerical
le=LabelEncoder()
df['Extracurricular Activities']=le.fit_transform(df['Extracurricular Activities'])
```

```python
df.head()
```

```python
x = df[["Hours Studied","Previous Scores", "Extracurricular Activities", "Sleep Hours", "Sample Question Papers Practiced"]]
x
```

```python
y=df['Performance Index']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
scaler=StandardScaler()
```

```python
x_train_scaled=scaler.fit_transform(x_train)
```

```python
x_test_scaled=scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
import pickle
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
model=LinearRegression()
```

```python
model.fit(x_train_scaled,y_train)
```

```python
y_pred=model.predict(x_test_scaled)
y_pred
```

```python
x_test_scaled
```

```python
model.predict([[ 1.53762007, -0.74634342,  1.00300451,  0.27280347, -0.21658553]])
```

```python
model.predict([[ 1.53762007,  1.69620333,  1.00300451,  0.27280347, -0.21658553]])
```

```python
mean_squared_error(y_test,y_pred)
```

```python
#accuracy
r2_score(y_test,y_pred)
```

```python
#save the model and all preprocessing step
with open('student_lr_final_model.pkl','wb') as file:
    pickle.dump((model,scaler,le),file)
```
